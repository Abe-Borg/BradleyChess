```python
def process_chess_data_corrected(file_path):
    with open(file_path, 'r') as file:
        games_data = []
        game_moves = {}
        game_labels = []  # List to keep track of game labels

        for line in file:
            if line.startswith('Game'):
                # When a new game starts, if the current game_moves dictionary is not empty,
                # append it to games_data list, add the game label to game_labels list, and start a new game_moves dictionary
                if game_moves:
                    games_data.append(game_moves)
                    game_moves = {}
                game_label = line.strip()  # Extract the game label directly from the line
                game_labels.append(game_label)
            else:
                # Check if line contains a comma, indicating it has a move and a score
                if ', ' in line:
                    move, score = line.strip().split(', ')
                    game_moves[move] = int(score)
        
        # Don't forget to add the last game_moves to the games_data list if not empty
        if game_moves:
            games_data.append(game_moves)
        
        # Convert the list of dictionaries into a DataFrame
        df = pd.DataFrame(games_data)
        df.index = game_labels  # Set the index to the game labels extracted from the file
        df.index.name = 'Game'
        
        return df
```

```python
# The file path where the text file is located
file_path = game_settings.est_q_vals_filepath_part_1
# Process the chess data with the corrected function
est_q_vals_df_part_1 = process_chess_data_corrected(file_path)

est_q_vals_df_part_1.head()
```
